```python
import torch

device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using device: {device}")

from diffusers import StableDiffusionPipeline

# Load DreamShaper model (No need to load the second model)
pipe = StableDiffusionPipeline.from_pretrained(
    "Lykon/dreamshaper-8",  # DreamShaper model path
    torch_dtype=torch.float16,
    variant="fp16",  # Use 16-bit precision for faster processing
).to(device)

# The model is now loaded and ready to generate images
```

```python
# Generate an image with a prompt
prompt = "A serene mountain landscape with a crystal-clear lake at sunrise"
image = pipe(prompt).images[0]

# Display the image
image.show()

# Save the image to a file
image.save("serene_landscape.png")
print("Image generated and saved successfully!")

# You can check and navigate the mounted path by
!ls /content/drive/MyDrive/
```
