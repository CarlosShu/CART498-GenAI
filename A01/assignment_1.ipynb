**LOAD STABLE DIFFUSION**

```python
from diffusers import StableDiffusionPipeline
import torch

model_id = "sd-legacy/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")
```

**PROMPT 1: MARCH TO WAR**

```python
prompt = "A black-and-white image of little army toy soldiers walking in single file atop a collaged background of fragmented clocks and jagged newspaper clippings."
image = pipe(prompt).images[0]

image.save("surreal_collage_1.png")
print("Image generated and saved successfully!")
```

**PROMPT 2: BATTLE OF WAR**

```python
prompt = "A black-and-white image of little army toy soldiers charging forward atop a collaged battlefield of fragmented clocks and jagged newspaper clippings, with an overturned military helmet lying to the side."
image = pipe(prompt).images[0]

image.save("surreal_collage_2.png")
print("Image generated and saved successfully!")
```

**PROMPT 3: DEVASTATION OF WAR**

```python
prompt = "A black-and-white image of little army toy soldiers lying motionless atop a collaged background of jagged newspaper clippings, with broken a clock hand scattered around them."
image = pipe(prompt).images[0]

image.save("surreal_collage_3.png")
print("Image generated and saved successfully!")
```
